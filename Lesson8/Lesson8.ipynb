```python
# é€™å€‹ç¨‹å¼ç¢¼æ¨¡æ“¬äº†ä¸€å€‹ gaming_utils.py æ¨¡çµ„

def calculate_xp_needed(current_level):
    """è¨ˆç®—åˆ°ä¸‹ä¸€ç´šæ‰€éœ€çš„ç¶“é©—å€¼"""
    return (current_level ** 2) * 100

def generate_random_loot():
    """ç‚ºç©å®¶ç”Ÿæˆéš¨æ©Ÿæˆ°åˆ©å“"""
    import random
    loot_types = ['âš”ï¸ åŠ', 'ğŸ›¡ï¸ ç›¾ç‰Œ', 'ğŸ’ å¯¶çŸ³', 'ğŸ¹ å¼“', 'ğŸ”® è—¥æ°´']
    rarity = random.choice(['æ™®é€š', 'ç¨€æœ‰', 'å²è©©', 'å‚³å¥‡'])
    item = random.choice(loot_types)
    return f"{rarity} {item}"

def player_stats(name, level, health, mana):
    """é¡¯ç¤ºç©å®¶çµ±è¨ˆè³‡æ–™"""
    return f"""
ğŸ® ç©å®¶çµ±è¨ˆ ğŸ®
åå­—ï¼š{name}
ç­‰ç´šï¼š{level}
è¡€é‡ï¼š{health} â¤ï¸
é­”åŠ›ï¼š{mana} ğŸ’™
    """

def calculate_damage(base_damage, level, critical_chance=0.1):
    """è¨ˆç®—æ”»æ“Šå‚·å®³"""
    import random
    damage = base_damage + (level * 2)

    # æš´æ“Šæª¢æŸ¥
    if random.random() < critical_chance:
        damage *= 2
        return damage, True  # æš´æ“Šï¼
    return damage, False

# æ¸¬è©¦æˆ‘å€‘çš„éŠæˆ²æ¨¡çµ„
player_name = "é¾æ®ºæ‰‹123"
player_level = 15
player_health = 100
player_mana = 50

print(player_stats(player_name, player_level, player_health, player_mana))
print(f"ä¸‹ä¸€ç´šæ‰€éœ€ç¶“é©—å€¼ï¼š{calculate_xp_needed(player_level)}")
print(f"ä½ ç™¼ç¾äº†ï¼š{generate_random_loot()}")

# æˆ°é¬¥æ¨¡æ“¬
damage, is_critical = calculate_damage(25, player_level)
if is_critical:
    print(f"ğŸ’¥ æš´æ“Šï¼é€ æˆ {damage} é»å‚·å®³ï¼")
else:
    print(f"âš”ï¸ é€ æˆ {damage} é»å‚·å®³")
```

## 4. çœŸå¯¦ä¸–ç•Œç¯„ä¾‹ï¼šç¤¾ç¾¤åª’é«”è²¼æ–‡åˆ†æå™¨ ğŸ“±
åƒæ•¸æ“šç§‘å­¸å®¶ä¸€æ¨£åˆ†æè²¼æ–‡ï¼

```python
import re
from datetime import datetime

def analyze_post(post_text):
    """åˆ†æç¤¾ç¾¤åª’é«”è²¼æ–‡"""

    # è¨ˆç®—å­—æ•¸
    word_count = len(post_text.split())

    # å°‹æ‰¾æ¨™ç±¤ï¼ˆ#ï¼‰å’ŒæåŠï¼ˆ@ï¼‰
    hashtags = re.findall(r'#\w+', post_text)
    mentions = re.findall(r'@\w+', post_text)

    # åµæ¸¬æƒ…ç·’
    positive_words = ['è¶…æ£’', 'å¤ªå¥½äº†', 'æ„›', 'é©šäºº', 'é–‹å¿ƒ', 'èˆˆå¥®', 'æ£’', 'å¥½']
    negative_words = ['è¨å­', 'ç³Ÿç³•', 'å¯æ€•', 'é›£é', 'ç”Ÿæ°£', 'å¤±æœ›', 'å£']

    positive_count = sum(1 for word in positive_words if word in post_text)
    negative_count = sum(1 for word in negative_words if word in post_text)

    if positive_count > negative_count:
        mood = "ğŸ˜Š æ­£é¢"
    elif negative_count > positive_count:
        mood = "ğŸ˜” è² é¢"
    else:
        mood = "ğŸ˜ ä¸­æ€§"

    return {
        'word_count': word_count,
        'hashtags': hashtags,
        'mentions': mentions,
        'mood': mood,
        'analyzed_at': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

# æ¸¬è©¦åˆ†æå™¨
test_posts = [
    "å‰›å®Œæˆæˆ‘çš„è¶…æ£’ Python å°ˆæ¡ˆï¼#ç¨‹å¼è¨­è¨ˆ #python @github",
    "ä»Šå¤©çš„è€ƒè©¦ç³Ÿç³•é€äº†ï¼Œæˆ‘è¨å­æ•¸å­¸ #è€ƒè©¦ #å£“åŠ›",
    "ä»Šå¤©å¤©æ°£ä¸éŒ¯ï¼Œå»äº†å…¬åœ’æ•£æ­¥ #æ—¥å¸¸ #ç”Ÿæ´»"
]

for i, post in enumerate(test_posts, 1):
    print(f"\nğŸ“Š è²¼æ–‡ {i} åˆ†æçµæœï¼š")
    print(f"å…§å®¹ï¼š{post}")
    analysis = analyze_post(post)
    for key, value in analysis.items():
        print(f"{key}: {value}")
    print("-" * 40)
```

## 5. äº’å‹•æŒ‘æˆ°ï¼šæˆç¸¾è¨ˆç®—ç³»çµ± ğŸ“š
å»ºç«‹ä¸€å€‹æ™ºèƒ½æˆç¸¾è¨ˆç®—å™¨ï¼

```python
def calculate_letter_grade(percentage):
    """å°‡ç™¾åˆ†æ¯”è½‰æ›ç‚ºå­—æ¯ç­‰ç´š"""
    if percentage >= 90:
        return "A+ ğŸ†"
    elif percentage >= 80:
        return "A ğŸŒŸ"
    elif percentage >= 70:
        return "B ğŸ‘"
    elif percentage >= 60:
        return "C ğŸ˜Š"
    else:
        return "F ğŸ“š (ç¹¼çºŒåŠªåŠ›ï¼)"

def calculate_gpa(grades):
    """è¨ˆç®— GPA"""
    grade_points = {'A+': 4.0, 'A': 4.0, 'B': 3.0, 'C': 2.0, 'F': 0.0}
    total_points = sum(grade_points.get(grade.split()[0], 0) for grade in grades)
    return total_points / len(grades) if grades else 0.0

def generate_report_card(student_name, subjects_grades):
    """ç”Ÿæˆæˆç¸¾å–®"""
    print(f"\nğŸ“ {student_name.upper()} çš„æˆç¸¾å–® ğŸ“")
    print("=" * 40)

    total_percentage = 0
    letter_grades = []

    for subject, percentage in subjects_grades.items():
        letter_grade = calculate_letter_grade(percentage)
        letter_grades.append(letter_grade)
        total_percentage += percentage
        print(f"{subject}ï¼š{percentage}% ({letter_grade})")

    average = total_percentage / len(subjects_grades)
    gpa = calculate_gpa(letter_grades)

    print("=" * 40)
    print(f"ç¸½å¹³å‡ï¼š{average:.1f}%")
    print(f"GPAï¼š{gpa:.2f}")
    print(f"æœ€çµ‚æˆç¸¾ï¼š{calculate_letter_grade(average)}")

    # çµ¦äºˆå»ºè­°
    if average >= 90:
        print("\nğŸŒŸ å„ªç§€è¡¨ç¾ï¼ç¹¼çºŒä¿æŒï¼")
    elif average >= 80:
        print("\nğŸ‘ å¾ˆå¥½çš„æˆç¸¾ï¼å†åŠªåŠ›ä¸€é»å°±èƒ½é”åˆ°å„ªç§€äº†ï¼")
    elif average >= 70:
        print("\nğŸ“– ä¸éŒ¯çš„è¡¨ç¾ï¼Œæœ‰é€²æ­¥ç©ºé–“ï¼")
    else:
        print("\nğŸ’ª éœ€è¦æ›´å¤šåŠªåŠ›ï¼Œä½†åˆ¥æ”¾æ£„ï¼")

# ç¤ºç¯„ä¸åŒå­¸ç”Ÿçš„æˆç¸¾
students_data = [
    ("ç‹å°æ˜", {"æ•¸å­¸": 85, "ç§‘å­¸": 92, "è‹±æ–‡": 78, "æ­·å²": 88, "ç¾è¡“": 95}),
    ("æå°è¯", {"æ•¸å­¸": 76, "ç§‘å­¸": 81, "è‹±æ–‡": 85, "æ­·å²": 79, "ç¾è¡“": 83}),
    ("é™³å°ç¾", {"æ•¸å­¸": 95, "ç§‘å­¸": 98, "è‹±æ–‡": 92, "æ­·å²": 96, "ç¾è¡“": 94})
]

for name, grades in students_data:
    generate_report_card(name, grades)
```

## 6. é€²éšæŒ‘æˆ°ï¼šå¯†ç¢¼å®‰å…¨æª¢æŸ¥å™¨ ğŸ”
å»ºç«‹ä¸€å€‹å°ˆæ¥­ç´šçš„å¯†ç¢¼å®‰å…¨ç³»çµ±ï¼

```python
import string
import random

def check_password_strength(password):
    """æª¢æŸ¥å¯†ç¢¼å¼·åº¦"""
    score = 0
    feedback = []

    # é•·åº¦æª¢æŸ¥
    if len(password) >= 12:
        score += 3
        feedback.append("âœ… é•·åº¦å¾ˆå¥½ï¼")
    elif len(password) >= 8:
        score += 2
        feedback.append("ğŸ‘ é•·åº¦ä¸éŒ¯")
    else:
        score += 0
        feedback.append("âŒ å¤ªçŸ­äº†ï¼è‡³å°‘ä½¿ç”¨ 8 å€‹å­—å…ƒ")

    # å­—å…ƒå¤šæ¨£æ€§æª¢æŸ¥
    if any(c.islower() for c in password):
        score += 1
        feedback.append("âœ… åŒ…å«å°å¯«å­—æ¯")
    else:
        feedback.append("âŒ åŠ å…¥å°å¯«å­—æ¯")

    if any(c.isupper() for c in password):
        score += 1
        feedback.append("âœ… åŒ…å«å¤§å¯«å­—æ¯")
    else:
        feedback.append("âŒ åŠ å…¥å¤§å¯«å­—æ¯")

    if any(c.isdigit() for c in password):
        score += 1
        feedback.append("âœ… åŒ…å«æ•¸å­—")
    else:
        feedback.append("âŒ åŠ å…¥æ•¸å­—")

    if any(c in string.punctuation for c in password):
        score += 1
        feedback.append("âœ… åŒ…å«ç‰¹æ®Šå­—å…ƒ")
    else:
        feedback.append("âŒ åŠ å…¥ç‰¹æ®Šå­—å…ƒ (!@#$%^&*)")

    # åˆ¤æ–·å¼·åº¦
    if score >= 6:
        strength = "ğŸ”’ è¶…å¼·"
    elif score >= 4:
        strength = "ğŸ” å¼·"
    elif score >= 2:
        strength = "ğŸ”‘ ä¸­ç­‰"
    else:
        strength = "ğŸš« å¼±"

    return strength, feedback, score

def generate_secure_password(length=12):
    """ç”Ÿæˆå®‰å…¨å¯†ç¢¼"""
    characters = string.ascii_letters + string.digits + "!@#$%^&*"
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

# æ¸¬è©¦ä¸åŒå¼·åº¦çš„å¯†ç¢¼
test_passwords = [
    "123456",
    "password",
    "Password123",
    "MySecurePassword123!",
    "Sup3r$3cur3P@ssw0rd2024!"
]

for password in test_passwords:
    strength, feedback, score = check_password_strength(password)
    print(f"\nğŸ” å¯†ç¢¼åˆ†æï¼š{password}")
    print(f"å¼·åº¦ï¼š{strength} (åˆ†æ•¸ï¼š{score}/7)")
    print("å»ºè­°ï¼š")
    for item in feedback:
        print(f"  {item}")
    print("-" * 40)

# ç”Ÿæˆä¸€äº›å®‰å…¨å¯†ç¢¼å»ºè­°
print("\nğŸ² å»ºè­°çš„å®‰å…¨å¯†ç¢¼ï¼š")
for i in range(3):
    secure_pwd = generate_secure_password()
    strength, _, score = check_password_strength(secure_pwd)
    print(f"{i+1}. {secure_pwd} ({strength})")
```
